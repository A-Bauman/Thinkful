### A script to clean up these files; scrubbing out dollar signs, cents, converting object types to integers or floats, etc.  In the case of the 'listings' file, removed a bunch of unneeded columns.

```python
import pandas as pd
pd.set_option('display.max_columns', 100)
```

```python
calendar_data = pd.read_csv('calendar.csv')
calendar_data.head(3)
```

### cleanup of strings, fill NaN's, convert data type

```python
calendar_data['price'] = calendar_data['price'].str.replace("$","")
calendar_data['price'] = calendar_data['price'].str.replace(",","")
calendar_data['price'] = calendar_data['price'].str.replace("[.][0-9]{2}","")
calendar_data['price'] = calendar_data['price'].fillna(0)
calendar_data['price'] = calendar_data['price'].astype(int)
calendar_data['available'] = calendar_data['available'].str.replace('t','1')
calendar_data['available'] = calendar_data['available'].str.replace('f','0')
calendar_data['available'] = calendar_data['available'].astype(int)
```

```python
calendar_data.to_csv('calendar_clean.csv', index=False)
```

### This was the 'dirtiest' file, and had a LOT of unnecessary columns; renamed a few just to ease typing and to compress the display a bit.

```python
listings = pd.read_csv('listings.csv', usecols=['id', 'neighbourhood', 'neighbourhood_cleansed', 'zipcode',
                       'latitude', 'longitude', 'property_type', 'room_type', 'accommodates', 'bathrooms',
                       'bedrooms', 'beds', 'bed_type', 'amenities', 'square_feet', 'price', 'weekly_price',
                       'security_deposit', 'cleaning_fee', 'calendar_updated', 'has_availability',
                       'number_of_reviews', 'review_scores_rating', 'review_scores_accuracy',
                       'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication',
                       'review_scores_location', 'review_scores_value', 'license', 'reviews_per_month'])

listings.rename(columns={'neighbourhood':'neighb', 'neighbourhood_cleansed':'neighb_cl', 'accommodates':'accom',
                         'zipcode':'zip', 'latitude':'lat', 'longitude':'long', 'bathrooms':'baths',
                         'has_availability':'has_avail', 'number_of_reviews':'num_revs',
                         'review_scores_ratings':'rev_ratings', 'review_scores_accuracy':'rev_acc',
                         'review_scores_cleanliness':'rev_clean', 'review_scores_checkin':'rev_checkin',
                         'review_scores_communication':'rev_comm', 'review_scores_location':'rev_loc',
                         'review_scores_value':'rev_val', 'reviews_per_month':'revs_per_mon'}, inplace=True)
listings.head()
```

```python
# Obvious mistake here, the weekly price is below daily; an extra zero present, will correct below
listings[listings['id'] == 4037590]
```

```python
listings['neighb'].unique()
```

```python
listings['neighb_cl'].unique()       #to see the difference with 'neighb'
listings['zip'].unique()             #zip codes as floats, that makes sense.....
```

```python
listings['zip'] = listings['zip'].fillna(0)
listings['zip'] = listings['zip'].astype(int)
```

```python
# Get the prices cleaned up
cols = ['price', 'weekly_price', 'security_deposit', 'cleaning_fee']
for col in cols:
    listings[col] = listings[col].str.replace("$","")
    listings[col] = listings[col].str.replace(",","")
    listings[col] = listings[col].str.replace("[.][0-9]{2}","")
    listings[col] = listings[col].fillna(0)
    listings[col] = listings[col].astype(int)

# Convert to numeric form
listings['has_avail'] = listings['has_avail'].str.replace('t','1')
listings['has_avail'] = listings['has_avail'].str.replace('f','0')
listings['has_avail'] = listings['has_avail'].astype(int)
listings['license'] = listings['license'].fillna(0)

listings.loc[(listings['price'] == 9000) & (listings['id'] == 4037590), 'price'] = 900
listings.loc[(listings['price'] == 8000) & (listings['id'] == 14887262), 'price'] = 80
```

```python
listings.to_csv('listings_clean.csv', index=False)
```
